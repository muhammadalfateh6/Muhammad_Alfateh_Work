```
import nltk
nltk.download('punkt')
import requests as rq
output = rq.get('https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt')
text= output.text
```

```
words_count = text.split(" ")
len(words_count)
```

```
from nltk.tokenize import word_tokenize
latest_text = word_tokenize(text)
```

```
latest_text
```

```
len(latest_text)
```

```
from nltk.probability import FreqDist
freq = FreqDist()

for words in latest_text:
  freq[words.lower()]+=1
freq
```

```
top_20 = freq.most_common(20)
top_20
```

```
freq['the']
```

**STEMMING**

```
from nltk.stem import PorterStemmer, LancasterStemmer
ps = PorterStemmer()
print(ps.stem('placed'))
print(ps.stem('placing'))
print(ps.stem('permission'))
```

```
for wrd in latest_text:
  stemmed_text = print(ps.stem(wrd))
```
